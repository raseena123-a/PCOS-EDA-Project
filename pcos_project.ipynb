```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("PCOS_data.csv")
```

```python
df.head()
```

```python

print("Shape of dataset:", df.shape)
```

```python
print("Column Names:")
print(df.columns)
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#data cleaning
# Remove leading/trailing spaces
df.columns = df.columns.str.strip()
```

```python
# Drop unwanted columns
df.drop(columns=["Sl. No", "Patient File No.", "Unnamed: 44"], inplace=True)
```

```python
# Rename target column for convenience
df.rename(columns={"PCOS (Y/N)": "PCOS"}, inplace=True)
```

```python
# Convert object columns to numeric
df["II    beta-HCG(mIU/mL)"] = pd.to_numeric(df["II    beta-HCG(mIU/mL)"], errors="coerce")
df["AMH(ng/mL)"] = pd.to_numeric(df["AMH(ng/mL)"], errors="coerce")
```

```python
# Fill missing values with median (safe choice)
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.head()
```

```python
#does age effect the pcos?
# Distribution of Age
sns.histplot(df["Age (yrs)"], kde=True)
plt.title("Age Distribution")
```

```python
# Distribution of BMI
sns.histplot(df["BMI"], kde=True)
plt.title("BMI Distribution")
```

```python
# Count of PCOS cases
sns.countplot(x="PCOS", data=df)
plt.title("PCOS Class Distribution")
```

```python
# BMI vs PCOS
sns.boxplot(x="PCOS", y="BMI", data=df)
```

```python
# Exercise habits vs PCOS
sns.countplot(x="Reg.Exercise(Y/N)", hue="PCOS", data=df)
```

```python
#How many women have PCOS?
# Count of PCOS and Non-PCOS cases
sns.countplot(x="PCOS", data=df, palette="Set2")
plt.title("Distribution of PCOS Cases")
plt.xlabel("PCOS (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.show()
```

```python
#Do PCOS patients have higher BMI?
plt.figure(figsize=(7, 5))
sns.boxplot(x="PCOS", y="BMI", data=df, palette="Set3")
plt.title("BMI Comparison Between PCOS and Non-PCOS Patients")
plt.xlabel("PCOS (0 = No, 1 = Yes)")
plt.ylabel("BMI")
plt.grid(True)
plt.show()
```

```python
#Is LH/FSH Ratio Higher in PCOS Patients?
# Group by PCOS status and calculate mean LH/FSH ratio
ratio_means = df.groupby("PCOS")["FSH/LH"].mean().reset_index()
plt.figure(figsize=(6, 4))
sns.barplot(x="PCOS", y="FSH/LH", data=ratio_means, palette="Set1")
plt.title("Average FSH/LH Ratio by PCOS Status")
plt.xlabel("PCOS (0 = No, 1 = Yes)")
plt.ylabel("Average FSH/LH Ratio")
plt.ylim(0, ratio_means["FSH/LH"].max() + 0.5)
plt.grid(True)
plt.show()
```

```python
#Do PCOS Patients Have More Irregular Cycles?
cycle_pc = pd.crosstab(df["Cycle(R/I)"], df["PCOS"])
cycle_pc.columns = ["No PCOS", "PCOS"]
cycle_pc.plot(kind="bar", stacked=True, figsize=(6, 5), colormap="Paired")
plt.title("Menstrual Cycle Type by PCOS Status")
plt.xlabel("Cycle Type (R = Regular, I = Irregular)")
plt.ylabel("Number of Patients")
plt.legend(title="Group")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#What is the average follicle count in PCOS vs Non-PCOS patients?
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x="PCOS", y="Follicle No. (L)", data=df, palette="Set2")
plt.title("Left Ovary Follicle Count by PCOS Status")
plt.xlabel("PCOS (0 = No, 1 = Yes)")
plt.ylabel("Follicle Count (Left)")
plt.subplot(1, 2, 2)
sns.boxplot(x="PCOS", y="Follicle No. (R)", data=df, palette="Set3")
plt.title("Right Ovary Follicle Count by PCOS Status")
plt.xlabel("PCOS (0 = No, 1 = Yes)")
plt.ylabel("Follicle Count (Right)")
plt.tight_layout()
plt.show()
```

```python
#Do lifestyle factors like fast food and exercise relate to PCOS?
#  Fast Food vs PCOS
fast_food_ct = pd.crosstab(df["Fast food (Y/N)"], df["PCOS"])
fast_food_ct.columns = ["No PCOS", "PCOS"]
fast_food_ct.plot(kind="bar", stacked=True, colormap="Pastel1", figsize=(6,4))
plt.title("Fast Food Habit by PCOS Status")
plt.xlabel("Fast Food Habit (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.legend(title="Group")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#  Regular Exercise vs PCOS
exercise_ct = pd.crosstab(df["Reg.Exercise(Y/N)"], df["PCOS"])
exercise_ct.columns = ["No PCOS", "PCOS"]
exercise_ct.plot(kind="bar", stacked=True, colormap="Pastel2", figsize=(6,4))
plt.title("Regular Exercise by PCOS Status")
plt.xlabel("Exercise Habit (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.legend(title="Group")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
